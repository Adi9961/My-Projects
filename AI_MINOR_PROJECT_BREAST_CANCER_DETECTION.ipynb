**#Getting dependent and independent variable from the dataset**

```
x=df.iloc[:,2:31].values
y=df.iloc[:,1].values
x
```

**#Splitting dataset into training and test set**

```
#Dataset splitted into 75% training and 25% testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
x_test.shape
```

**#Scaling of all values between 0 and 1**

```
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

**#Applying ANN and finding the best values for the hyperparameteres**

```
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
def build_classifier():
  classifier=Sequential()
  classifier.add(Dense(units=16,kernel_initializer='uniform',activation='relu',input_dim=29))
  classifier.add(Dense(units=16,kernel_initializer='uniform',activation='relu'))
  classifier.add(Dense(units=1,kernel_initializer='uniform',activation='sigmoid'))
  classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
  return classifier
classifier=KerasClassifier(build_fn=build_classifier)
parameters={'batch_size':[10,20,30],'epochs':[100,200,300]}
grid_search=GridSearchCV(estimator=classifier,param_grid=parameters,scoring='accuracy',cv=5,n_jobs=-1)
grid_search=grid_search.fit(x_train,y_train)
best_parameters=grid_search.best_param_
```

**#Applying ANN with optimized values of hyperparametres**

```
from keras.models import Sequential
from keras.layers import Dense
```

```
classifier=Sequential()
```

```
classifier.add(Dense(units=16,kernel_initializer='uniform',activation='relu',input_dim=29))
classifier.add(Dense(units=16,kernel_initializer='uniform',activation='relu'))
classifier.add(Dense(units=1,kernel_initializer='uniform',activation='sigmoid'))
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
history=classifier.fit(x_train,y_train,epochs=100,batch_size=29,validation_data=(x_test,y_test))
```

```
accuracy=history.history['accuracy']
val_accuracy=history.history['val_accuracy']
loss=history.history['loss']
val_loss=history.history['val_loss']
epochs=range(len(accuracy))
plt.plot(epochs,accuracy,'ro',label='Training Accuracy')
plt.plot(epochs,val_accuracy,'b',label='Validation Accuracy')
plt.title('Training and Validation accuracy')
plt.legend()
```

```
plt.plot(epochs,loss,'ro',label='Training Accuracy')
plt.plot(epochs,val_loss,'b',label='Validation Accuracy')
plt.title('Training and Validation loss')
plt.legend()
plt.show()
```

**#Prediction on test data**

```
pred=classifier.predict(x_test)
pred=(pred>0.5)
```

**#Confusion matrix**

```
#Test model accuracy on test data on confusion matrix
from sklearn.metrics import confusion_matrix
AI=confusion_matrix(y_test,pred)
TP=AI[0][0]
TN=AI[1][1]
FN=AI[1][0]
FP=AI[0][1]
print(AI)
ac=round((TP+TN)/(TP+TN+FN+FP)*100,2)
print('Testing Accuracy=',ac)
print()
```

**#Heatmap for visualization the accuracy of ANN on test data**

```
sns.heatmap(AI,annot=True)
```

**#Hence the final concluded accuracy**

```
print("Finally the accuracy is:",ac)
```
